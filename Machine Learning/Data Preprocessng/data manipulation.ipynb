```python
# Libraries
import pandas as pd
```

```python
#import data
df = pd.read_csv('dataset/Salaries.csv', low_memory=False)
df.head()
```

```python
# Remove the none columns
df = df.drop(['Benefits', 'Notes'], axis=1)
df.head()
```

```python
# total salary cost from year 2011 to 2014
mean_year = df.groupby('Year').mean()['TotalPayBenefits']
mean_year
```

```python
# Top earning employee
top_salary = df.groupby('EmployeeName').sum()['TotalPayBenefits']
top_salary.sort_values(axis=0)
```
